```
# Clean Disease and Symptom Names
def clean_data(data):
    data_list = []
    data_name = data.replace('^','_').split('_')
    n = 1
    for names in data_name:
        if (n % 2 == 0):
            data_list.append(names)
        n += 1
    return data_list
```

```
disease_list = []
disease_symptom_dict = defaultdict(list)
disease_symptom_count = {}
count = 0

for idx, row in data.iterrows():
    
    # Get the Disease Names
    if (row['Disease'] !="\xc2\xa0") and (row['Disease'] != ""):
        disease = row['Disease']
        disease_list = clean_data(data=disease)
        count = row['Count of Disease Occurrence']

    # Get the Symptoms Corresponding to Diseases
    if (row['Symptom'] !="\xc2\xa0") and (row['Symptom'] != ""):
        symptom = row['Symptom']
        symptom_list = clean_data(data=symptom)
        for d in disease_list:
            for s in symptom_list:
                disease_symptom_dict[d].append(s)
            disease_symptom_count[d] = count
```

```
# Writing the data in a csv file

with open("/content/drive/My Drive/Symptoms_clean.csv","w") as csvfile:
    writer = csv.writer(csvfile)
    for key,values in disease_symptom_dict.items():
        for v in values:
            #key = str.encode(key)
            key = str.encode(key).decode('utf-8')
            #.strip()
            #v = v.encode('utf-8').strip()
            #v = str.encode(v)
            writer.writerow([key,v,disease_symptom_count[key]])
```

```
def _fixCount(df):
  l = df.Disease.unique()
  for i in range(0, 1):
    df1 = df.loc[df['Disease'] == l[i]]
    df1[1:].Count = df1[1:].Count.replace(df1[1:].Count.unique(), 0)
  for i in range(1, len(l)):
    df2 = df.loc[df['Disease'] == l[i]]
    df2[1:].Count = df2[1:].Count.replace(df2[1:].Count.unique(), 0)
    df1 = pd.concat([df1, df2])
  return df1
```

```
def _getCountData(arr):
  d = np.zeros((arr[0].shape[0]), dtype=int)
  for i in range(0, len(arr)):
    count = 0
    #print(i)
    while (count < arr[i][-1:]):
      #print(count)
      d = np.vstack((d, arr[i]))
      count += 1
  return d[1:]
```

Data Augmentation

```
def _dataAugmentation(x, y):
  i = 0
  _data = np.zeros((1, x.shape[1]), dtype=int)
  _labels = np.zeros(1, dtype=int)
  while(i < 1000):
    _data = np.vstack((_data, x))
    _labels = np.hstack((_labels, y))
    i += 1

  return _data[1:], _labels[1:]
```

```
x = np.asarray(x)
y = np.asarray(y)
trainX, trainY = _dataAugmentation(x, y)
trainY = trainY.reshape(-1, 1)
data = np.hstack((trainX, trainY))
np.random.shuffle(data)
```

```
def _getSplits(arr):
    data = []
    labels = []
    for i in range(0, len(arr)):
        a = arr[i][:-1]
        data.append(a)
        
        b = arr[i][-1:]
        labels.append(b)
        
    X = np.asarray(data)
    y = np.asarray(labels)
    return X, y
```

```
x, y = _getSplits(data)
```
