```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
from google.colab import drive
drive.mount("/content/drive")
```

```
data_drugs = pd.read_csv('drive/My Drive/255_Project/DataSet/DrugReview_processed.csv')
data_symptoms = pd.read_csv('drive/My Drive/255_Project/DataSet/Symptoms_processed.csv')
```

```
data_drugs.head()
```

```
data_drugs.drop(['uniqueID', 'date'], axis = 1, inplace=True)
```

```
data_symptoms.head()
```

```
data_drugs = data_drugs.rename(columns={"condition": "Disease"})
```

```
data_drugs.dropna(axis = 0, inplace=True)
data_drugs.sort_values(by = ['Disease'], ascending= False , inplace= True)
data_drugs.head()
```

```
data_drugs['Disease'].unique()
```

```
data_symptoms.dropna(axis = 0, inplace=True)
data_symptoms.sort_values(by = ['Disease'], ascending= False , inplace= True)
data_symptoms.head()
```

```
data_drugs['Disease'] = data_drugs['Disease'].str.lower()
data_symptoms['Disease'] = data_symptoms['Disease'].str.lower()
```

```
merged_data = pd.merge(data_drugs, data_symptoms, on='Disease')
```

```
merged_data.columns = ['Drug', 'Disease', 'Review','Rating' ,'UsefulCount', 'Symptoms']
```
